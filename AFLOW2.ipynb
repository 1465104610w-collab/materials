```python
import requests
get_data=[]
repond=requests.get('https://oqmd.org/oqmdapi/formationenergy?fields=name,entry_id,spacegroup,ntypes,band_gap,delta_e&filter=element_set=Fe,S')
txt=repond.json()
for i in txt['data']:
    get_data.append(i)
    print(i)
print(txt)
print(get_data)
print(len(get_data))
```

```python
print(get_data)
```

```python
len(get_data)
```

```python
next=txt['links']['next']
while(next):
    repond=requests.get(next)
    txt=repond.json()
    for i in txt['data']:
        get_data.append(i)
    next=txt['links']['next'] 
    print(next)
print(len(get_data))
```

```python
oqmd_dict = {get_data[i]['entry_id']:[get_data[i]['name'],get_data[i]['spacegroup'],get_data[i]['ntypes'],get_data[i]['band_gap'],get_data[i]['delta_e']] for i in range(len(get_data))}
print(oqmd_dict)
```

```python
import pandas as pd
df=pd.DataFrame.from_dict(oqmd_dict,orient="index")
df.head()
```
