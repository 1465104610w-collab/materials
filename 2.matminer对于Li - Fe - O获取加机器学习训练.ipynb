```python
import pymatgen
```

```python
from pymatgen.ext.matproj import MPRester
```

```python
from pymatgen.ext.matproj import MPRester
with MPRester ("DxTbWWBsxEID35nq") as mpr:
    criteria = {"elements":{"$all":["Li","Fe","O"]},"nelements":3}
    properties = ['material_id', 'pretty_formula','energy_per_atom','spacegroup','structure','density','volume']
    shortC = (mpr.query(criteria,properties))
    print(shortC)
```

```python
import pandas as pd
dfC = pd.DataFrame(shortC)
dfC
```

```python
from matminer.featurizers.structure import XRDPowderPattern
```

```python
xrd = XRDPowderPattern(two_theta_range=[0,300])
xrd.feature_labels()
```

```python
dfC = xrd.featurize_dataframe(dfC,"structure")
dfC
```

```python
y = dfC['energy_per_atom'].values
y
```

```python
X = dfC.drop(['material_id','structure','spacegroup', 'pretty_formula'],axis = 1)
X
```

# 随机森林训练

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.7,random_state=30)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, random_state=1)
rf.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt

y_pred = rf.predict(X_test)
plt.plot(y_test,'ro')
plt.plot(y_pred,'bo')

mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
R2 = r2_score(y_pred,y_test)
print('training R2 = {:.3f} ev/Atom'.format(R2))
print('training mae = {:.3f} ev/Atom'.format(mae))
print('training mse = {:.3f} ev/Atom'.format(mse))
print('training RMSE = {:.3f} ev/Atom'.format(np.sqrt(mse)))
```

# GBR训练

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.7,random_state=30)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
GBR = GradientBoostingRegressor(random_state=30)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
parameters = {'loss':['ls','lad','huber','quantile'],
             'min_samples_leaf':[1,2,3,4,5],
             'alpha':[0.1,0.3,0.6,0.9]}
```

```python
grid = GridSearchCV(estimator=GBR, param_grid=parameters,n_jobs= -1,cv=10) 
GBR_model = grid.fit(X_train,y_train)
```

```python
pred_1 = GBR_model.predict(X_train)
pred_1
```

```python
grid.best_score_
```

```python
best_parameters = grid.best_params_
best_parameters
```

```python
gbr_preds_1 = GBR_model.predict(X_test)
gbr_preds_1
```

```python
mae = mean_absolute_error(y_test,gbr_preds_1)
R2 = r2_score(gbr_preds_1,y_test)
mse = mean_squared_error(gbr_preds_1,y_test)
print('training R2 = {:.3f} ev/Atom'.format(R2))
print('training mae = {:.3f} ev/Atom'.format(mae))
print('training mse = {:.3f} ev/Atom'.format(mse))
print('training RMSE = {:.3f} ev/Atom'.format(np.sqrt(mse)))
```
