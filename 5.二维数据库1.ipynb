```python
import requests

url = "https://materialsdb.cn/2dsdb/index.html?"
headers = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36 Edg/111.0.1661.41'
}
response = requests.get(url=url,headers=headers)
print(response.text)
```

```python
import re
html_data = re.findall('<li><p><a class="reference external" href="(.*?).html">(.*?)</a></p></li>',response.text)
html_data
```

# 获取整个界面的内容

```python
import os

filename = "materials\\"
if not os.path.exists(filename):
    os.mkdir(filename)
```

```python
for num_id in html_data:
    print(num_id[0])
```

```python
for num_id in html_data:
    num = num_id[0]
    materials_url = f'https://materialsdb.cn/2dsdb/{num}.html'
    print(materials_url)
```

```python
for num_id in html_data:
    num = num_id[0]
    graph_url = f'https://materialsdb.cn/_images/BAND_LDOS-{num}.png'
    graph_content = requests.get(url=graph_url,headers=headers).content
    with open("materials\\" + num_id[0] + '.png',mode='wb')as f:
        f.write(graph_content)
```
