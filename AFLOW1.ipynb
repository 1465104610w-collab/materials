```python
import json
import requests
cont=[]
gap=[]
density=[]
energyperatom=[]
API='https://aflowlib.duke.edu/'
for i in range(len(txt)):
    ADRESS=txt[i]['aurl'][18:]
    print(ADRESS)
    que_density=requests.get(API+ADRESS+'/?density').text
    density.append(que_density)
    que_gap=requests.get(API+ADRESS+'/?Egap').text
    gap.append(que_gap)
    que_energy=requests.get(API+ADRESS+'/?energy_atom').text
    energyperatom.append(que_energy)    
    que_cont=requests.get(API+ADRESS+'/CONTCAR.relax').text
    cont.append(que_cont)     
    print(i)
print(cont[0])
print(density[0])
print(gap[0])
print(energyperatom[0])
```
