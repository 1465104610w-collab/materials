```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlretrieve
import os
```

```python
url = 'https://proceedings.neurips.cc/paper/2021'
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36 Edg/110.0.1587.69'
}
response = requests.get(url=url, headers=headers)
print(response.text)
```

# 创建BeautifulSoup对象

```python
str_html = requests.get(url)
soup = BeautifulSoup(str_html.text, 'lxml')
print(soup)
```

```python
data = soup.select('body > div.container-fluid > div > ul > li > a')
print(data)
```

```python
print(len(data))
```

```python
downloadUrl_list = []
for i,j in enumerate(data):#i:index , j:name
    downloadUrl = "https://proceedings.neurips.cc" + data[i].get('href')
    downloadUrl_list.append(downloadUrl)
print(downloadUrl_list)
```

```python
title_list = []
for item in data:
    title_list.append([item.get_text(), item.get('href')])
print(title_list)
```

```python
name = ['title', 'link']
test = pd.DataFrame(columns=name, data=title_list)
test
test.to_csv("title_link.csv",index=False)
```

```python
def pdf_url(url):
    str_subhtml = requests.get(url)
    soup_1 = BeautifulSoup(str_subhtml.text, 'lxml')
    sub_data = soup_1.select('body > div.container-fluid > div > div > a:nth-child(4)')
    downloadUrl = 'https://proceedings.neurips.cc' + sub_data[0].get('href')
    return downloadUrl
```

```python
pdf_url_link = downloadUrl_list[0]
link = pdf_url(pdf_url_link)
```
